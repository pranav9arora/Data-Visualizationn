```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
custom_dict = {}

for x in set(df['Age']):
    if x == "All":
        custom_dict[x] = 110
    elif x == "100+":
        custom_dict[x] = 100
    else:
        custom_dict[x] = int(x.split("--")[0])
        

        
df = pd.read_csv('prac.csv')
df['Sorter'] = df['Age'].map(custom_dict)
df.sort(columns=['Area', 'Sorter'],inplace=True)
table = pd.pivot_table(df, values='Ratio', index=['Area', 'Age'], columns=['Year'], fill_value=0)
```

```python
tables = [pd.pivot_table(df.loc[df['Area'] == country], values='Ratio', index=['Sorter'], columns=['Year'], fill_value=0) for country in sorted(list(set(df['Area'])))]
```

```python
ticks = []

for x in set(df['Sorter']):
    if x < 100:
        ticks.append(str(x)+"--"+str(x+4))
    elif x==100:
        ticks.append("100+")
    else:
        ticks.append("All")
```

```python
req = input("Enter Country Name: ")

if req not in sorted(list(set(df['Area']))):
    print("Invalid Input")
else:
    for i in range(len(sorted(list(set(df['Area']))))):
        if sorted(list(set(df['Area'])))[i] == req:
            ax = sns.heatmap(tables[i])
            ax.set_yticklabels(reversed(ticks))
            plt.show()
            break
```
